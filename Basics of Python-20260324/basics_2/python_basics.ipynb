# Basics of Python

```python
# Data-types:
# -----------

integer = 10

floaty = 10.0

stringy = "Ingolstadt"
stringy_2 = "10"

complex = 5+2j
print(complex.real)
print(complex.imag)

print(integer + float(stringy_2))

print("Sum=", 2+3)
print("Multiply=", 2*3)
print("Division=",5/2)
print("Division=",5//2)
print("Modulus=",5%2)
print("Exponent=",5**2)
```

```python
# Booleans:
# ---------

booly = bool("False")

print(booly)

booly_2 = bool(" ")
print(booly_2)

booly_3 = bool(0.0+0.0j)
print(booly_3)
```

```python
# Complex data types:
# -------------------
"""
Note: Keys should be immutable data types
"""

listy = [1,2,3,4,5, "True", True]

tuply = (1,2,3,4,5)

dictionary = {(1,2,3): "Chid"}

print(dictionary[(1,2,3)])
```

```python
# Functions:
# ----------

# Value returning function
def sample_func(x: int, 
                y: int)->int:
    return x**y

# Keyword arguments:
# ------------------
result = sample_func(y=3, x=2)
print(result)

# Void function
def sample_func(x: int, 
                y: int)->int:
    print(x**y)

result = sample_func(y=3, x=2)
print(result)
```

```python
# Object oriented programming:
# ----------------------------

# Parent class
class Vehicle:
    def __init__(self, num_wheels):
        self.num_wheels = num_wheels

    # Method:
    def drive(self):
        print("Driving")


car_1 = Vehicle(num_wheels=4)

car_2 = Vehicle(num_wheels=6)

print(car_1.num_wheels)

car_2.num_wheels = 8

print(car_2.num_wheels)
```

```python
# Child class:
class Motorbike(Vehicle):
    def __init__(self, num_wheels, colour):
        super().__init__(num_wheels)
        self.colour = colour

    def wheelie(self):
        print("I am being reckless! 😝")


ducati = Motorbike(num_wheels=2,
                   colour="Red")


ducati.drive()
print(ducati.num_wheels)
print(ducati.colour)
```

# Basics of Numpy

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
my_image = np.array(Image.open("dog.jpg"))

plt.imshow(my_image);
```

```python
print("Shape: ", my_image.shape)
```

```python
# Slicing arrays: RGB channels
red_channel = my_image[:, :, 0]
green_channel = my_image[:, :, 1]
blue_channel = my_image[:, :, 2]

fig, axes = plt.subplots(1, 3, figsize=(12, 4))

axes[0].imshow(red_channel, cmap="Reds")
axes[0].set_title("Red channel")
axes[0].axis("off")

axes[1].imshow(green_channel, cmap="Greens")
axes[1].set_title("Green channel")
axes[1].axis("off")

axes[2].imshow(blue_channel, cmap="Blues")
axes[2].set_title("Blue channel")
axes[2].axis("off")

plt.tight_layout()
plt.show()
```
